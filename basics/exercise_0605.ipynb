```python
a = "Life is too short, You need Python" 
len(a)
```

```python
a = "Life is too short, You need Python"
a[3]
a[-2]
a[0:4] # 슬라이싱기법에서 마지막 인덱스는 포함되지 않는다. 즉, 0,1,2,3 까지만 포함된다.
a[19:] # 19번 인덱스부터 끝까지
a[:17] # 처음부터 17번 인덱스까지 (17번 인덱스는 포함되지 않는다.)
```

```python
>>> a = "20230331Rainy"
>>> date = a[:8]
>>> weather = a[8:]
>>> date
'20230331'
>>> weather
'Rainy'
```
